```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from keras.optimizers import RMSprop
from keras.regularizers import l1

# Initialize the model
alexnet = Sequential()

# Add convolutional layers
alexnet.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)))
alexnet.add(Conv2D(32, (3,3), padding='same', activation='relu'))
alexnet.add(MaxPooling2D((2,2), strides=(2,2), padding='same'))
alexnet.add(Dropout(0.5))  # Dropout layer

alexnet.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
alexnet.add(Conv2D(64, (3,3), padding='same', activation='relu'))
alexnet.add(MaxPooling2D((2,2), strides=(2,2), padding='same'))
alexnet.add(Dropout(0.5))  # Dropout layer

alexnet.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
alexnet.add(Conv2D(128, (3,3), padding='same', activation='relu'))
alexnet.add(Conv2D(128, (3,3), padding='same', activation='relu'))
alexnet.add(MaxPooling2D((2,2), strides=(2,2), padding='same'))
alexnet.add(Dropout(0.5))  # Dropout layer

# Flatten layer
alexnet.add(Flatten())

# Fully connected layers
alexnet.add(Dense(128, activation="relu"))
alexnet.add(Dropout(0.8))  # Dropout layer
alexnet.add(BatchNormalization())  # Batch normalization layer

# Output layer
alexnet.add(Dense(1, kernel_regularizer=l1(0.01), activation="linear"))

# Compile the model
learning_rate = 0.0001
opti = RMSprop(lr=learning_rate)
alexnet.compile(optimizer=opti, loss='hinge', metrics=['accuracy'])

# Train the model
history = alexnet.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=200, batch_size=batch_size)
```
